```python
df.drop_duplicates(inplace=True)
```

```python
# mark species to numerical values
df['species'] = df['species'].astype('category').cat.codes
```

```python
# Assume df is your DataFrame
sns.pairplot(df, diag_kind='hist', hue = 'species', palette='Set2')

# Customize the plot (optional)
plt.suptitle('Pair Plot of Dataset', y=1.02)  # To Add a title
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
corr = df.corr()

# Plot the heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')

# Customize the plot
plt.title('Correlation Heatmap')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

#features and target variables
x = df.drop('species', axis=1)
y = df['species']

#split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
#to import logisticregression class from sklearn
from sklearn.linear_model import LogisticRegression

#to initialize the logisticregression model
logistic_model = LogisticRegression()

#to fit the model to the trainig data(x_train,y_train)
logistic_model.fit(X_train, y_train)
```

```python
y_pred_logistic = logistic_model.predict(X_test)
print(y_pred_logistic)
```

```python
from sklearn.metrics import accuracy_score

# Assuming you have already made predictions using the logistic regression model
y_pred_logistic = logistic_model.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred_logistic)

# Print the accuracy
print(f"Accuracy of the logistic regression model: {accuracy:.2f}")
```
